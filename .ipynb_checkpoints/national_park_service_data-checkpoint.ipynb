```python
states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", 
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
```

```python
national_park_state = []
national_park_name = []
national_park_designation = []
national_park_code = []
```

```python
full_url = base_url + 'parks?&limit=490&stateCode=' + 'CA'+'&sort=&sort=' + '&api_key='+ national_park_key_2 

response=requests.get(full_url).json()
response
```

```python
for state in states:
    full_url = base_url + 'parks?&limit=100&stateCode=' + state + '&api_key='+ national_park_key_2
    response=requests.get(full_url).json()
    for i in range(int(response['total'])):
        print(i)
        print(state)
        print(response['data'][i]['fullName'])
        print(response['data'][i]['designation'])
        print(response['data'][i]['parkCode'])
        national_park_bool = ((response['data'][i]['designation'] == "National and State Parks") |(response['data'][i]['designation'] == "National Parks") | (response['data'][i]['designation'] == "National Park") | (response['data'][i]['designation'] == "National Park & Preserve"))
        if national_park_bool:
            national_park_state.append(state)
            national_park_name.append(response['data'][i]['fullName'])
            national_park_designation.append(response['data'][i]['designation'])
            national_park_code.append(response['data'][i]['parkCode'])
```

```python
national_park_df = pd.DataFrame({'State' : national_park_state, 'Park Designation': national_park_designation, 'national park full name': national_park_name, 'Park Code' : national_park_code})
```

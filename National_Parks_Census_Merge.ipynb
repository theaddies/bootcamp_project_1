```python
park_path = "../bootcamp_project_1/Resources/national_parks_clean_with_zip.csv"
census_path = "../bootcamp_project_1/Resources/dataclean1.csv"

parks_df = pd.read_csv(park_path)
census_df = pd.read_csv(census_path)
```

```python
merge_ready_parks_df = parks_df.rename(columns={'postalCode': 'Zipcode'})

new_zip = []
for code in merge_ready_parks_df['Zipcode']:    
    if ('-' in code) == True:
        new_zip.append(code.split('-', 1)[0])

    else:
        new_zip.append(code)
        
merge_ready_parks_df['Zipcode'] = new_zip
print(merge_ready_parks_df['Zipcode'])
```

```python
#merge_ready_parks_df['Zipcode'] = merge_ready_parks_df['Zipcode'].astype(str).astype(int)
#merge_ready_parks_df.dtypes

census_df['Zipcode'] = census_df['Zipcode'].astype(int).astype(str)
census_df.dtypes
```

```python
combined_park_census_df = pd.merge(merge_ready_parks_df, census_df, how='outer', on = 'Zipcode')

combined_park_census_df = combined_park_census_df.drop(columns = ['Unnamed: 0_x'])
combined_park_census_df = combined_park_census_df.drop(columns = ['Unnamed: 0_y'])
combined_park_census_df = combined_park_census_df.dropna()

combined_park_census_df
#combined_park_census_df.isnull().sum()
```

```python
combined_park_census_df.to_csv('NationalPark_Census_Merge.csv')
```
